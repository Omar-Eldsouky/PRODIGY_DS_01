```python
import pandas as pd

# Load the data from the CSV files
file_female_population = '/home/kirito/Data Science with prodigy/API_SP.POP.TOTL.FE.IN_DS2_en_csv_v2_1589620/API_SP.POP.TOTL.FE.IN_DS2_en_csv_v2_1589620.csv'
file_male_population = '/home/kirito/Data Science with prodigy/API_SP.POP.TOTL.MA.IN_DS2_en_csv_v2_1597510/API_SP.POP.TOTL.MA.IN_DS2_en_csv_v2_1597510.csv'
file_total_population = '/home/kirito/Data Science with prodigy/API_SP.POP.TOTL_DS2_en_csv_v2_1991674/API_SP.POP.TOTL_DS2_en_csv_v2_1991674.csv'

# Read the CSV files
df_female = pd.read_csv(file_female_population, skiprows=4)
df_male = pd.read_csv(file_male_population, skiprows=4)
df_total = pd.read_csv(file_total_population, skiprows=4)

# Display the first few rows of each DataFrame to understand the structure
df_female.head(), df_male.head(), df_total.head()
```

```python
# Extract relevant columns and rename for clarity
df_female_population = df_female.melt(id_vars=['Country Name', 'Country Code'], 
                                      var_name='Year', value_name='Female Population')
df_male_population = df_male.melt(id_vars=['Country Name', 'Country Code'], 
                                  var_name='Year', value_name='Male Population')
df_total_population = df_total.melt(id_vars=['Country Name', 'Country Code'], 
                                    var_name='Year', value_name='Total Population')

# Convert 'Year' to numeric type
df_female_population['Year'] = pd.to_numeric(df_female_population['Year'], errors='coerce')
df_male_population['Year'] = pd.to_numeric(df_male_population['Year'], errors='coerce')
df_total_population['Year'] = pd.to_numeric(df_total_population['Year'], errors='coerce')

# Merge DataFrames on 'Country Name', 'Country Code', and 'Year'
df_merged = df_female_population.merge(df_male_population, 
                                       on=['Country Name', 'Country Code', 'Year'], how='inner')
df_merged = df_merged.merge(df_total_population, 
                            on=['Country Name', 'Country Code', 'Year'], how='inner')

# Display the first few rows of the merged DataFrame
df_merged.head()
```

```python
# Remove rows where 'Year' is not a valid numeric value
df_female_population = df_female_population.dropna(subset=['Year'])
df_male_population = df_male_population.dropna(subset=['Year'])
df_total_population = df_total_population.dropna(subset=['Year'])

# Convert 'Year' to int type after dropping NaNs
df_female_population['Year'] = df_female_population['Year'].astype(int)
df_male_population['Year'] = df_male_population['Year'].astype(int)
df_total_population['Year'] = df_total_population['Year'].astype(int)

# Merge DataFrames again on 'Country Name', 'Country Code', and 'Year'
df_merged = df_female_population.merge(df_male_population, 
                                       on=['Country Name', 'Country Code', 'Year'], how='inner')
df_merged = df_merged.merge(df_total_population, 
                            on=['Country Name', 'Country Code', 'Year'], how='inner')

# Display the first few rows of the cleaned and merged DataFrame
df_merged.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_female_population = '/home/kirito/Data Science with prodigy/API_SP.POP.TOTL.FE.IN_DS2_en_csv_v2_1589620/API_SP.POP.TOTL.FE.IN_DS2_en_csv_v2_1589620.csv'
file_male_population = '/home/kirito/Data Science with prodigy/API_SP.POP.TOTL.MA.IN_DS2_en_csv_v2_1597510/API_SP.POP.TOTL.MA.IN_DS2_en_csv_v2_1597510.csv'
file_total_population = '/home/kirito/Data Science with prodigy/API_SP.POP.TOTL_DS2_en_csv_v2_1991674/API_SP.POP.TOTL_DS2_en_csv_v2_1991674.csv'

df_female = pd.read_csv(file_female_population, skiprows=4)
df_male = pd.read_csv(file_male_population, skiprows=4)

# Remove unnecessary columns
df_female = df_female.drop(columns=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'])
df_male = df_male.drop(columns=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'])

# Sum the populations for each year to get the global population
female_population_global = df_female.sum()
male_population_global = df_male.sum()

# Create a DataFrame for plotting
years = female_population_global.index
global_population = pd.DataFrame({
    'Year': years,
    'Female Population': female_population_global.values,
    'Male Population': male_population_global.values
})

# Plot the bar chart
plt.figure(figsize=(14, 8))
plt.bar(global_population['Year'], global_population['Female Population'], label='Female Population')
plt.bar(global_population['Year'], global_population['Male Population'], label='Male Population', bottom=global_population['Female Population'])
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Global Population Distribution by Gender for Each Year')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
